```python
#Clearing the dataset and normalising the dataset
dataset1 = dataset1.select_dtypes(exclude=['object'])            #excluding those columns which are of type object because they are categorical data
dataset1 =dataset1[~dataset1.isin([np.nan, np.inf, -np.inf]).any(1)]  #checing for the nan, inf, -inf in the dataset
```

```python
#Split the dataset into feature and target set

X = dataset1.copy()

#dropping all the features which are either irrelevant for the classifcation or having only one value
X.drop(list(X.filter(regex='cpu_guest')), axis = 1, inplace = True)
X.drop(list(X.filter(regex='cpu_guest_nice')), axis = 1, inplace = True)
X.drop(list(X.filter(regex='diskio_sda1_disk_name')), axis = 1, inplace = True)
X.drop(list(X.filter(regex='diskio_sda1_key')), axis = 1, inplace = True)
X.drop(list(X.filter(regex='diskio_sda_disk_name')), axis = 1, inplace = True)
X.drop(list(X.filter(regex='diskio_sda_key')), axis = 1, inplace = True)
X.drop(list(X.filter(regex='fs_/_fs_type')), axis = 1, inplace = True)
X.drop(list(X.filter(regex='fs_/_key')), axis = 1, inplace = True)
X.drop(list(X.filter(regex='fs_/_mnt_point')), axis = 1, inplace = True)
X.drop(list(X.filter(regex='network_lo_interface_name')), axis = 1, inplace = True)
X.drop(list(X.filter(regex='network_lo_key')), axis = 1, inplace = True)
X.drop(list(X.filter(regex='percpu_0_cpu_number')), axis = 1, inplace = True)
X.drop(list(X.filter(regex='percpu_0_guest')), axis = 1, inplace = True)
X.drop(list(X.filter(regex='percpu_0_guest_nice')), axis = 1, inplace = True)
X.drop(list(X.filter(regex='percpu_0_irq')), axis = 1, inplace = True)
X.drop(list(X.filter(regex='percpu_0_key')), axis = 1, inplace = True)
X.drop(list(X.filter(regex='percpu_0_steal')), axis = 1, inplace = True)
X.drop(list(X.filter(regex='system_hr_name')), axis = 1, inplace = True)
X.drop(list(X.filter(regex='system_linux_distro')), axis = 1, inplace = True)
X.drop(list(X.filter(regex='system_os_name')), axis = 1, inplace = True)
X.drop(list(X.filter(regex='system_platform')), axis = 1, inplace = True)

#dropping the timestamp feature becasuse time constraints conflicts in building the model
X.drop(list(X.filter(regex = 'timestamp')), axis = 1, inplace = True) 

#cpu_guest
#cpu_guest_nice
#diskio_sda1_disk_name
#diskio_sda1_key
#diskio_sda_disk_name
#diskio_sda_key

#fs_/_fs_type
#fs_/_key
#fs_/_mnt_point

#network_lo_interface_name
#network_lo_key

#percpu_0_cpu_number
#percpu_0_guest
#percpu_0_guest_nice
#percpu_0_irq
#percpu_0_key
#percpu_0_steal

#system_hr_name
#system_linux_distro
#system_os_name
#system_platform

#timestamp
```
